```python
import sys, torch
sys.path.insert(0, 'src')
from sentiment.data import load_rotten_tomatoes
from sentiment.train_bert import train_bert

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('device:', device)
Xtr, ytr, Xte, yte = load_rotten_tomatoes()

model, tokenizer, metrics = train_bert(Xtr, ytr, Xte, yte, epochs=2)
print(metrics)
```

```python
# Save + push to HF Hub. Token comes from Colab Secrets (key icon, left sidebar):
# Name: HF_TOKEN, Value: your HF write token -> enable notebook access.
from google.colab import userdata
from huggingface_hub import HfApi
token = userdata.get('HF_TOKEN')
repo = 'ashishjangra284/distilbert-rotten-tomatoes'
model.push_to_hub(repo, token=token)
tokenizer.push_to_hub(repo, token=token)
import json
with open('bert_metrics.json','w') as f: json.dump(metrics, f, indent=2)
HfApi(token=token).upload_file(path_or_fileobj='bert_metrics.json', path_in_repo='eval_metrics.json', repo_id=repo)
print('done -> https://huggingface.co/' + repo)
```

Weights published — the demo Space can now load `ashishjangra284/distilbert-rotten-tomatoes` directly.
